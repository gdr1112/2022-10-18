```python
train_input, test_input, train_target, test_target = train_test_split(
    perch_full, perch_weight, random_state = 42
)
train_input.shape, test_input.shape, train_target.shape, test_target.shape
```

## Scikit-Learn Transformer
- 데이터의 특성을 변환시키는 여러 함수들


```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(include_bias = False)

#2와 3으로 구성된 샘플적용
poly.fit([[2, 3]])

print(poly.transform([[2, 3]]))
```

무게 = a * 길이 + b * 높이+ c* 두께 + d(상수)

```python
for idx in range(2, 6):
  poly = PolynomialFeatures(degree = idx ,include_bias = False)
  poly.fit(train_input)

  train_poly = poly.transform(train_input)
  test_poly = poly.transform(test_input)

  lr = LinearRegression()
  lr.fit(train_poly, train_target)

  print("Degree: ", idx)
  print("차원의 개수 :" , train_poly.shape)
  print(lr.score(train_poly, train_target))
  print(lr.score(test_poly, test_target))
  print('-'*30)
```

```python
test_poly = poly.transform(test_input)
test_poly.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()  # 표준 점수
ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
#릿지 회귀 , 라쏘 회귀
#릿지 : 계쑤를 제곱한 값을 기준으로 규제를 적용
#라쏘 : 계수의 절댓값을 기준으로 규제를 적용ㄴ
```

```python
from sklearn.linear_model import Ridge
```

```python
import matplotlib.pyplot as plt 

train_score = []
test_score = []

alpha_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
for alpha in alpha_values:

  # 릿지 모델 
  ridge = Ridge(alpha = alpha)
  ridge.fit(train_scaled, train_target)

  # 훈련 점수와 테스트 점수를 저장
  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))

# print(train_score)
# print(test_score)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(np.log10(alpha_values), train_score)
ax.plot(np.log10(alpha_values), test_score)
ax.set_xlabel('alpha') # 릿지회귀 규제의 강도
ax.set_ylabel('R^2') # 결정계수
plt.show()
```

```python
ridge = Ridge(alpha = 0.1)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```
